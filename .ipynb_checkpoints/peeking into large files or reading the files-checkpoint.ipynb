```python
# Peeking into Large Files
#!/usr/local/bin/python
import sys
import string
print('What file would you like to read?\n')
line = input()
```

```python
line = line.rstrip()
infile = open (line, 'r')
print ('\n')
while(1):
    for iterations in range(40):
        print(infile.readline().rstrip())
    print('Type QUIT if you want to quit. Otherwise press any key\n')
    response = sys.stdin.readline().rstrip()
    if (response == 'QUIT'):
        break
infile.close()
exit
```
